```python
import matplotlib.pyplot as plt 
import matplotlib
import numpy as np 
import IPython
from IPython import display 

display.set_matplotlib_formats('svg')

plt.rcParams['figure.figsize'] = (10, 10)
#plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
fig = plt.figure(figsize=(10, 5))
fig.clf()  #清除所有的axes
ax = fig.gca()
for i in range(100):
    #ax.clear()
    ax.cla()
    ax.plot(x[:i], y[:i])
    #display.display(plt.gcf())
    display.display(fig)  #展示图片
    display.clear_output(wait=True) #清除输出
```

```python
1.设置栅格
(1)使用pyplot api命令

打开栅格:plt.grid(true)

设置栅格格式：plt.grid(color='r', linestyle='--', linewidth=1,alpha=0.3)

 

(2)使用axes类面向对象命令

#同时设置两坐标轴上的栅格线

ax.grid(color='r', linestyle='--', linewidth=1,alpha=0.3)  

#设置X坐标轴上(垂直方向)的栅格线

ax.xaxis.grid(color='r', linestyle='--', linewidth=1,alpha=0.3)

#设置Y坐标轴上(水平方向)的栅格线

ax.yaxis.grid(color='r', linestyle='--', linewidth=1,alpha=0.3)

2.设置axes脊柱(坐标系)
 

(1)去掉脊柱(坐标系)

ax.spines['top'].set_visible(False) #去掉上边框

ax.spines['bottom'].set_visible(False) #去掉下边框

ax.spines['left'].set_visible(False) #去掉左边框

ax.spines['right'].set_visible(False) #去掉右边框


(2)移动脊柱

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))


(3)设置边框线颜色

ax = plt.gca() # 获取当前的axes

ax.spines['right'].set_color('blue')ax.spines['top'].set_color('none')

(4)设置边框线宽ax1.spines['left'].set_linewidth(5)

(5)设置边框线型ax.spines['left'].set_linestyle('--')
 

3.设置背景颜色

(1)设置figure背景颜色

facecolor：背景颜色  edgecolor：边框颜色

plt.figure(facecolor='blue',edgecolor='black')

或

fig=plt.gcf()
fig.set_facecolor('green')
 

(2)设置axes背景颜色

a = plt.axes([.65, .6, .2, .2], facecolor='k')  #pyplot api命令-黑色背景

或者

ax1=plt.gca()

ax1.patch.set_facecolor("gray")            #设置ax1区域背景颜色               

ax1.patch.set_alpha(0.5)                      #设置ax1区域背景颜色透明度       

(3)修改matplotlib默认参数


plt.rcParams['axes.facecolor']='red'

plt.rcParams['savefig.facecolor']='red'
```

```python
#美化一下
config = {
    "font.family":'serif',
    "font.size": 15,
    "mathtext.fontset":'stix',
    "font.serif": ['SimSun'],
}
plt.rcParams.update(config)

fig = plt.figure(figsize=(10, 5))
fig.clf()  #清除所有的axes
ax = fig.gca()
ax.set_xlim([0, 10])

for i in range(100):

    ax.cla()  #清除改axe

    #清除上轴和右轴
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['top'].set_color('none')
    ax.yaxis.set_ticks_position('left')
    ax.spines['right'].set_color('none')
    #移动x轴和y轴
    ax.spines['bottom'].set_position(('data',0))
    #设置ticks
    ax.set_xticks([np.pi/2, np.pi, np.pi*3/2, 2*np.pi, 5*np.pi/2, 3*np.pi])
    ticklabels = ax.set_xticklabels([r'$\frac{\pi}{2}$',r'$\pi$',r'$\frac{3\pi}{2}$',r'$2\pi$',r'$\frac{5\pi}{2}$',r'$3\pi$'], fontsize=15)
    ticklabels[2].set_color('red')
    #ticklabels[1].set_math_fontfamily('stix')#'dejavusans', 'dejavuserif', 'cm', 'stix', 'stixsans', 'custom'
    #math_family = ['dejavusans', 'dejavuserif', 'cm', 'stix', 'stixsans', 'custom']
    #for i in range(len(math_family)):
        #ticklabels[i].set_math_fontfamily(math_family[i])
    ticklabels[0].set_bbox(rectprops=dict(facecolor='green', edgecolor='black', fill=True, alpha=0.2))
    ticklabels[3].set_visible(False)

    ax.set_xlim([0, 10])
    ax.set_ylim([-1, 1])
    ax.plot(x[:i], y[:i])
    ax.grid(color='green', alpha=0.3)
    ylabel = ax.set_ylabel('$\sin{x}$')
    #ylabel.set_rotation(0)
    #ylabel.set_position((30, 0.4))
    display.display(fig)  #展示图片
    display.clear_output(wait=True) #清除输出
```
